# 丢弃法

丢弃法（dropout），也是一种**正则化**手段。也就意味着只在训练中有效，在验证和测试中不进行 dropout 运算。

对一个多层感知机，训练时在输入数据处进行 dropout ，等价于在训练数据中引入噪声，已被证明为正则化方法。（具有输入噪声的训练等价于 Tikhonov正则化）

更常见的是在隐藏层的输出进行 dropout，当前理论认为是一种正则化方法。具体操作是**无偏的将一些输出随机置 0**。

在标准暂退法正则化中，通过按保留（未丢弃）的节点的分数进行规范化来消除每一层的偏差。
换言之，每个 中间活性值$h$ 以 *暂退概率*$p$ 由 随机变量$h'$ 替换，如下所示：

$$
\begin{aligned}
h' =
\begin{cases}
    0 & \text{ 概率为 } p \\
    \frac{h}{1-p} & \text{ 其他情况}
\end{cases}
\end{aligned}
$$

根据此模型的设计，其期望值保持不变，即 $E[h'] = h$。

**暂退概率p 即为 dropout 的超参数**，值越大在训练时 dropout 越强（正则化越强），一般取0.1 0.5 0.9
