# 权重衰退

权重衰退（weight decay），又可以被称为 L2正则化，与岭回归（使用 MSE方均根误差 和 L2正则化 的模型）类似。

本质上是通过限制权重的大小（权向量的模长的平方）来控制模型容量（模型复杂度）。

作用**是为了控制（在验证集上的）过拟合，而不是为了更好的找到（对训练集的）最优解**。

权重衰退通过超参数$\lambda$来控制，值越大则对权向量的控制越强。经验上取1e-4 1e-3 1e-2即可（本身对模型性能影响不大，仅控制过拟合用）。
